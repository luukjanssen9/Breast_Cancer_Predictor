```python
# import libraries
import numpy as np
import seaborn as sb
import plotly as ply
import matplotlib.pyplot as plt
import sklearn as skl
import pandas as pd
```

```python
# import csv file
breast_data = pd.read_csv("../datasets/brca.csv")
```

```python
breast_data = breast_data.drop(columns=["Unnamed: 0"])
```

```python
# one-hot encoding of y
breast_y_ohc = pd.get_dummies(breast_data["y"])
print(breast_y_ohc)
```

```python
breast_data = breast_data.join(breast_y_ohc)
breast_data = breast_data.drop(columns=["y"])
breast_data.head()
```

```python
# plot distributions of data

itr = 0
for item in breast_data.items():
    sb.displot(data=breast_data, x=item[0])
    itr += 1
```

```python
# plot benign data's distributions
benign_data = breast_data[breast_data['B'] == True]
benign_data = benign_data.drop(columns=['B','M'])

itr = 0
for item in benign_data.items():
    sb.displot(data=benign_data, x=item[0])
    itr += 1
```

```python
# plot malignant data's distributions
malignant_data = breast_data[breast_data['M'] == True]
malignant_data = malignant_data.drop(columns=['B','M'])

itr = 0
for item in malignant_data.items():
    sb.displot(data=malignant_data, x=item[0])
    itr += 1
```
